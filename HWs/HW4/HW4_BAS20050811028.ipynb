## SkLearn Fit

```python
from sklearn import linear_model as lm
from sklearn import preprocessing as pp
```

```python
def sklearn_fit_predict(X, Y, deg = 1):
    temp = lm.LinearRegression()
    poly = pp.PolynomialFeatures(degree=deg,include_bias=False)
    temp.fit(poly.fit_transform(X),Y)
    return temp.predict(poly.fit_transform(X))
```

```python
# Corn
predict_NDVI_one_sklearn_corn   = sklearn_fit_predict(days,corn_NDVI, deg = 1)
predict_NDVI_two_sklearn_corn   = sklearn_fit_predict(days,corn_NDVI, deg = 2)
predict_NDVI_three_sklearn_corn = sklearn_fit_predict(days,corn_NDVI, deg = 3)
predict_NDVI_four_sklearn_corn  = sklearn_fit_predict(days,corn_NDVI, deg = 4)
predict_NDVI_five_sklearn_corn  = sklearn_fit_predict(days,corn_NDVI, deg = 5)
# Cotton
predict_NDVI_one_sklearn_cotton   = sklearn_fit_predict(days,cotton_NDVI, deg = 1)
predict_NDVI_two_sklearn_cotton   = sklearn_fit_predict(days,cotton_NDVI, deg = 2)
predict_NDVI_three_sklearn_cotton = sklearn_fit_predict(days,cotton_NDVI, deg = 3)
predict_NDVI_four_sklearn_cotton  = sklearn_fit_predict(days,cotton_NDVI, deg = 4)
predict_NDVI_five_sklearn_cotton  = sklearn_fit_predict(days,cotton_NDVI, deg = 5)
```

## Manual Fit

```python
def manual_fit_predict(X, Y, deg = 1):
    poly = pp.PolynomialFeatures(degree=deg,include_bias=True)
    poly_feature = poly.fit_transform(X)
    cofs = np.linalg.solve(poly_feature.T.dot(poly_feature),poly_feature.T.dot(Y))
    return poly_feature.dot(cofs)
```

```python
# Corn
predict_NDVI_one_manual_corn   = manual_fit_predict(days,corn_NDVI, deg = 1)
predict_NDVI_two_manual_corn   = manual_fit_predict(days,corn_NDVI, deg = 2)
predict_NDVI_three_manual_corn = manual_fit_predict(days,corn_NDVI, deg = 3)
predict_NDVI_four_manual_corn  = manual_fit_predict(days,corn_NDVI, deg = 4)
predict_NDVI_five_manual_corn  = manual_fit_predict(days,corn_NDVI, deg = 5)
# Cotton
predict_NDVI_one_manual_cotton   = manual_fit_predict(days,cotton_NDVI, deg = 1)
predict_NDVI_two_manual_cotton   = manual_fit_predict(days,cotton_NDVI, deg = 2)
predict_NDVI_three_manual_cotton = manual_fit_predict(days,cotton_NDVI, deg = 3)
predict_NDVI_four_manual_cotton  = manual_fit_predict(days,cotton_NDVI, deg = 4)
predict_NDVI_five_manual_cotton  = manual_fit_predict(days,cotton_NDVI, deg = 5)
```

# SECTION TWO

```python
def ridge_fit(X, Y, deg = 1, a=0.0):
    temp = lm.Ridge(alpha=a)
    poly = pp.PolynomialFeatures(degree=deg,include_bias=False)
    temp.fit(poly.fit_transform(X),Y)
    return temp.coef_

def lasso_fit(X, Y, deg = 1, a=0.0):
    temp = lm.Lasso(alpha=a)
    poly = pp.PolynomialFeatures(degree=deg,include_bias=False)
    temp.fit(poly.fit_transform(X),Y)
    return temp.coef_
```

```python
predict_NDVI_corn = sklearn_fit_predict(days,corn_NDVI, deg = 3)
predict_NDVI_cotton  = sklearn_fit_predict(days,cotton_NDVI, deg = 4)
days_NEW = pp.StandardScaler().fit_transform(days)
```

```python
n_alphas = 50
regularization_alphas = np.logspace(-8, 5, n_alphas, base = 10.0)

corn_ridge_coefs = []
cotton_ridge_coefs = []
corn_lasso_coefs = []
cotton_lasso_coefs = []
for alph in regularization_alphas:
    corn_ridge_coefs.append(ridge_fit(days_NEW,corn_NDVI,deg=4,a=alph).reshape(4))
    cotton_ridge_coefs.append(ridge_fit(days_NEW,cotton_NDVI,deg=4,a=alph).reshape(4))
    corn_lasso_coefs.append(lasso_fit(days_NEW,corn_NDVI,deg=4,a=alph).reshape(4))
    cotton_lasso_coefs.append(lasso_fit(days_NEW,cotton_NDVI,deg=4,a=alph).reshape(4))
```

```python
plt.plot(regularization_alphas, np.array(cotton_ridge_coefs),"g",label=["Cotton a1","Cotton a2","Cotton a3","Cotton a4"])
plt.plot(regularization_alphas, np.array(corn_ridge_coefs),"y",label=["Corn a1","Corn a2","Corn a3","Corn a4"])
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.xscale('log')
plt.yscale('symlog')
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(regularization_alphas, np.array(cotton_lasso_coefs),"g",label=["Cotton a1","Cotton a2","Cotton a3","Cotton a4"])
plt.plot(regularization_alphas, np.array(corn_lasso_coefs),"y",label=["Corn a1","Corn a2","Corn a3","Corn a4"])
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Lasso coefficients as a function of the regularization')
plt.xscale('log')
plt.yscale('symlog')
plt.legend()
plt.grid()
plt.show()
```

We observe the 4th coefficient of corn starts very close to 0 and starts collapsing before every other coefficient.  
Which means the contribution of the 4th coefficiet is very low.  
**Meaning we can get a relatively accurate prediction with only 3 coefficients.**  
\
As for cotton, We observe that all of it's coefficients provide relatively high on the end result.  
And they collapse roughly at the same time.  
**Which means we cannot express cotton using any degree less than 4.**  
\
Note that We used only 4 coefficients because that was the highest degree we could enter before the regularization became divergent.
